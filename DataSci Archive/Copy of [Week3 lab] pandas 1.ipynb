```python
import pandas as pd
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 1000)
```

```python
df_exams = pd.read_csv('/content/StudentsPerformance.csv')
print(df_exams)
```

```python
df_exams['average score'] = (df_exams['math score'] * 0.5) + \
                            (df_exams['reading score'] * 0.25) + \
                            (df_exams['writing score'] * 0.25)
print(df_exams)
```

```python
def assign_grade(score):
    if score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'

df_exams['grade'] = df_exams['average score'].apply(assign_grade)
print(df_exams)
```

```python
num_A = df_exams[df_exams['grade'] == 'A'].shape[0]
perc_A = (num_A / df_exams.shape[0]) * 100
print(num_A,perc_A)
```

```python
male_A = df_exams[(df_exams['grade']=='A') &  (df_exams['gender'] == 'male')].shape[0]
print(male_A)
male_Num = df_exams[df_exams['gender'] == 'male'].shape[0]
perc_Male_A = (male_A/male_Num)*100
print(male_A,perc_Male_A)
```

```python
A_asc = df_exams[(df_exams['grade']=='A') & (df_exams['parental level of education'] == "associate's degree")].shape[0]
A_High = df_exams[(df_exams['grade']=='A') & (df_exams['parental level of education'] == "high school")].shape[0]
print(A_asc>A_High)
```

```python
A_students_per_education_level = df_exams[df_exams['grade'] == 'A']['parental level of education'].value_counts()
print(A_students_per_education_level)
```

```python
avg_Math = df_exams.groupby('parental level of education')['math score'].mean()
print(avg_Math)
```
