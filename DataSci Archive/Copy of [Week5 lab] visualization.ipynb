Load the Titanic dataset from Seaborn and create a visualization to answer the following question.

Q1) how different in terms of fare between males and females?

Q2) Do we see more surviving people in females and the first class?

Q3) Do we spot any fare difference compared to age and sex? Can we spot an outlier fares in the plot?

Q4) We want to see the number of males and females in each age group(bin). (please set a bin step to 10. E.g. 0-10,10-20,20-30, ...)



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 1000)

data = sns.load_dataset("titanic")
print(data)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='fare', y='sex', data=data, palette={'male': '#347DC1', 'female': '#CC6594'})
plt.title('Fare Distribution by Gender')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='class', hue='sex', data=data[data['survived'] == 1] ,palette={'male': '#A9CEEA', 'female': '#CC6594'})
plt.title('Survival Count by Class and Gender')
plt.show()
```

```python
sns.jointplot(x='age', y='fare', hue='sex', data=data,palette={'male': 'blue', 'female': 'red'})
plt.title('Fare vs Age Colored by Gender')
plt.show()
```

```python
data['age_group'] = pd.cut(data['age'], bins=range(0, 81, 10))
sns.countplot(x='age_group', hue='sex', data=data,palette={'male': 'blue', 'female': 'red'})
plt.title('Number of males and females in each age group')
plt.show()
```
