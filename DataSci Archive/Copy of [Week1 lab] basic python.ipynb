# Question 1. Create a function to find Min, Max, Mean, and Median on a list.
Input(parameters):
1. name of the function(String) and
2. List of values

Output:
1. Metric value

PS. sort vs sorted [link text](https://www.geeksforgeeks.org/python-difference-between-sorted-and-sort/)

```python
import statistics
vals = [7,14,81,56,9,17,43]


def getMetric(metric,ListVal):
   ListVal.sort()
   if metric == "mean":
      return statistics.mean(ListVal)
   elif metric == "median":
      return statistics.median(ListVal)
   return

mean = getMetric("mean",vals)
med = getMetric("median",vals)
print(mean)
print(med)
```


# Question 2. Count the number of words appearing in text.
*   read text file from sampleText.txt
*   Count number of each word by the appearance in text
*   the result should show like below

Result
1. sed 5650
2. in 4748
3. amet 4594
4. sit 4541
5. id 3853
6. ...

PS. To sort Dict: [article](https://thispointer.com/sort-a-dictionary-by-value-in-python-in-descending-ascending-order/)

```python
import re

inPath = "/content/sampleText.txt"

with open(inPath) as f: #read as one long string
    contents = f.read()

words = contents.split()
for w in words:
    w = re.sub(r'[^A-Za-z]', '', w)

print(words[:10])
```

```python
dictw = {}
for w in words:
  if w not in dictw:
    dictw[w] = 1
  else:
    dictw[w] +=1
dictw = dict(sorted(dictw.items(),key = lambda item:item[1],reverse = True))
```
