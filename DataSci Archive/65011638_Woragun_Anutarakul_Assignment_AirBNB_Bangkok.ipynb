```python
#Q3
from collections import Counter
all_amenities = [item for sublist in df_List['amenities'] for item in sublist]
unique_amenities = set(all_amenities)
num_unique_amenities = len(unique_amenities)

amenities_count = Counter(all_amenities)
top_three_amenities = amenities_count.most_common(3)
num_unique_amenities, top_three_amenities
```

```python
#Q4
sep_2022 = df_Carl[df_Carl['date'].str.startswith('2022-09')]
print(len(sep_2022))
print(sep_2022)

```

```python
#Q5
average_prices = df_Carl.groupby('listing_id')['price'].mean()
print(average_prices.max())
```

```python
#Q6
price_difference = df_Carl.groupby('listing_id')['price'].agg(['min', 'max'])
price_difference['difference'] = price_difference['max'] - price_difference['min']
listings_large_diff = price_difference[price_difference['difference'] > 10000]
num_listings_large_diff = len(listings_large_diff)
num_listings_large_diff
```

```python
#Q7
from scipy.stats import zscore

price_difference['z_score'] = zscore(price_difference['difference'])
outliers = price_difference[price_difference['z_score'].abs() > 3]
num_outliers = len(outliers)
outlier_hosts = df_List[df_List['id'].isin(outliers.index)]['host_name']
num_outliers, outlier_hosts
```

```python
#Q8
print(outlier_hosts.unique())
```

```python
#Q9
listings_per_malicious_host = df_List[df_List['id'].isin(outliers.index)].groupby('host_name').size()
print(listings_per_malicious_host)
malicious_host = df_List[df_List['id'].isin(outliers.index)]
print(malicious_host)
```

```python
#Remove Malicious

clean_List = df_List[~df_List['host_name'].isin(malicious_host['host_name'])]
#print(clean_List)
#Q10
top_3_districts = clean_List['neighbourhood_cleansed'].value_counts().head(3)
print(top_3_districts)

print(df_List)
print(clean_List)
#print(outliers)
```

```python
#Q12
Parthumwan_2B = clean_List[(clean_List['beds'] == 2) &(clean_List['neighbourhood_cleansed'] == 'Parthum Wan')]
print(len(Parthumwan_2B))
```

```python
#Q13
parthum_wan_two_bedrooms_ids = Parthumwan_2B['id']
parthum_wan_prices = df_Carl[df_Carl['listing_id'].isin(parthum_wan_two_bedrooms_ids)]
min_P = parthum_wan_prices['price'].min()
max_P = parthum_wan_prices['price'].max()
mean_P = parthum_wan_prices['price'].mean()
print("Min",min_P,"Max",max_P,"Mean",round(mean_P,2))
```

```python
#Q14
parthum_wan_high_rating = clean_List[(clean_List['beds'] == 2) &
                                        (clean_List['neighbourhood_cleansed'] == "Parthum Wan") &
                                        (clean_List['review_scores_rating'] >= 3)]

print(len(parthum_wan_high_rating))
```
