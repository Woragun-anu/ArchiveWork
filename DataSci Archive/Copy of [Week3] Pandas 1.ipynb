## How to create dataframe.  

1.   From an array - a) numpy array and b) list
2.   From dictionary
3.   From CSV files
4.   From JSON - tutorial ([a](https://datatofish.com/load-json-pandas-dataframe/))





```python
# 1a) creating an array
import numpy as np
data = np.array([[1, 4], [2, 5], [3, 6]])
# creating a dataframe
df = pd.DataFrame(data, index=['1', '2', '3'],
                 columns=['col1', 'col2'])
print(df)
```

```python
# 1b) creating an array with list shape
data = [[1, 4], [2, 5], [3, 6]]

# creating a dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'],
                 columns=['col1', 'col2'])
print(df)

df = pd.DataFrame(data)
print(df)
```

```python
# 2)  From dictionary
states = ["California", "Texas", "Florida", "New York","vegas"]
population = [39613493, 29730311, 21944577, 19299981, None] #None

dict_states = {'States': states, 'Population': population}
df_population = pd.DataFrame(dict_states)
print(df_population)
```

### indexing desired rows and cols
1. .loc[row,col] OR [] is primarily label based. Use boolean as index condition. If the condition is true, pandas will return subset of df back.
2. .iloc[row,col] is primarily integer position based.   


```python
print(df_exams.columns)

# select rows where gender equal male.
print(df_exams['gender']=="male")
print(df_exams[df_exams['gender']=="male"])

print(df_exams.loc[df_exams['gender']=="male"])
print(df_exams.loc[df_exams['gender']=="male"])
```

```python
print(df_exams.loc[df_exams['math score']>80])
print(df_exams.loc[df_exams['math score']>80][["gender","parental level of education"]])
print(df_exams.loc[df_exams['math score']>80,["gender","parental level of education"]])
```

Select with multiple conditions  


*   use () to seperate condition
*   use & for AND condition
*   use | for OR condition
* use ~ for NOT operation


```python
selectDf = df_exams[(df_exams['math score'] > 70) & (df_exams['gender']=="male")]
print(selectDf)
```

```python
selectDf = df_exams[(df_exams['math score'] > 70) & ~(df_exams['gender']=="male")]
print(selectDf)
```

.iloc - Select by position .iloc[row position, col position]

```python
# : use as from : to.
# , use as individual position
print(df_exams.iloc[0:3, [0,2]])

print(df_exams.iloc[0:3, 0:2])
```

# Operations on dataframe

###Adding a new column with a scalar value

```python
print(df_exams)
```

```python
df_exams['language score'] = 70
print(df_exams)
```

###Adding a new column with an array

```python
langScore= list(range(0, 1000))
print(len(langScore))
df_exams['language score'] =langScore #must have the same length to the original dataframe
print(df_exams)
```

### Adding new column from other columnss

```python
df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'])/3
print(df_exams)
```

### Operations in rows

```python
print(df_exams['math score'].count())
print(df_exams['math score'].mean())
print(df_exams['math score'].std())
print(df_exams['math score'].max())
print(df_exams['math score'].min())

print(df_exams["math score"].describe())
```

### Reset Index  
*   reset a dataframe with default numerical value.
*   a new row index as a range of numbers starting at 0.
*   Previous index will be added as a new col



pandas-dataframe-reset-index.webp

```python
copyDf = df_exams[(df_exams['math score'] > 70) & (df_exams['gender']=="male")]
print(copyDf)

copyDf = copyDf.reset_index(drop=True)  #drop=True
print(copyDf)

```

### Modifying value on dataframe based on condition

\*** **use .loc to make change on existing dataframe** \***




```python
print(df_exams)
df_exams = df_exams.drop('average',axis='columns')
print(df_exams)
```

```python
print(df_exams[df_exams['math score'] > 79])

df_exams.loc[df_exams['math score'] > 79, 'reading score'] = 100
df_exams[df_exams['math score'] > 79]["writing score"] = 100 # trigger warning that set on a copy of a slice from a DataFrame

print(df_exams[df_exams['math score'] > 79])
```

### Apply function to all rows in columns

```python
print(df_exams)
df_exams['math score'] = df_exams['math score'].apply(lambda x: x + 1)  # x is df_exams['math score']
print(df_exams)
```

```python
df_exams['sum_score'] = df_exams.apply(lambda x: x["math score"] + x["reading score"]+x["writing score"] , axis=1)
print(df_exams)
```

```python
def plusOne(x):
    y = x+1
    return y

df_exams['math score'] = df_exams.apply(lambda x: plusOne(x["math score"]), axis=1)
print(df_exams)
```

### Sort dataframe

```python
df_exams = df_exams.sort_values(['math score'], ascending=False) #false large >> small
print(df_exams)

df_exams = df_exams.sort_values(['math score', 'writing score'], ascending=False)
print(df_exams)
```
