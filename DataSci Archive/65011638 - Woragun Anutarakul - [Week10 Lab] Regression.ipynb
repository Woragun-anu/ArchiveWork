```python

df = pd.read_csv("/content/CarPrice_Dataset_GDSC_article.csv")
```

```python
from sklearn.preprocessing import OneHotEncoder
df['CarName'] = df['CarName'].astype('category')
df['fueltype'] = df['fueltype'].astype('category')
df['aspiration'] = df['aspiration'].astype('category')
df['carbody'] = df['carbody'].astype('category')
df['drivewheel'] = df['drivewheel'].astype('category')
df['enginelocation'] = df['enginelocation'].astype('category')
df['enginetype'] = df['enginetype'].astype('category')
df['fuelsystem'] = df['fuelsystem'].astype('category')
df['doornumber'] = df['doornumber'].astype('category')
df['cylindernumber'] = df['cylindernumber'].astype('category')

df['CarName'] = df['CarName'].cat.codes
df['fueltype'] = df['fueltype'].cat.codes
df['aspiration'] = df['aspiration'].cat.codes
df['carbody'] = df['carbody'].cat.codes
df['drivewheel'] = df['drivewheel'].cat.codes
df['enginelocation'] = df['enginelocation'].cat.codes
df['enginetype'] = df['enginetype'].cat.codes
df['fuelsystem'] = df['fuelsystem'].cat.codes
df['doornumber'] = df['doornumber'].cat.codes
df['cylindernumber'] = df['cylindernumber'].cat.codes
enc = OneHotEncoder()
enc_data = pd.DataFrame(enc.fit_transform(
    df[['CarName', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'fuelsystem', 'doornumber', 'cylindernumber']]).toarray())
New_df = df.join(enc_data)
```

```python
X=df.drop(['price'], axis=1)
y=df['price']
print(X)
print(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=42)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_predLR = lr.predict(X_test)
# print(y_test)
# print(y_predLR)

resultDf = pd.DataFrame({'y_test': y_test, 'y_predict':y_predLR})
print(resultDf)
```

```python
mean_absolute_error(y_test, y_predLR)
```

## Task 3
Task 3: Instead of using train_test_split with 0.1 ratios, implement K-fold validation and train the regression model with the same settings.

* Use 5-fold validation and report the average MAE score.

[ Using Kfold article.](https://www.askpython.com/python/examples/k-fold-cross-validation)






```python
from sklearn.model_selection import KFold
df = pd.read_csv("/content/CarPrice_Dataset_GDSC_article.csv")
from sklearn.preprocessing import OneHotEncoder
df['CarName'] = df['CarName'].astype('category')
df['fueltype'] = df['fueltype'].astype('category')
df['aspiration'] = df['aspiration'].astype('category')
df['carbody'] = df['carbody'].astype('category')
df['drivewheel'] = df['drivewheel'].astype('category')
df['enginelocation'] = df['enginelocation'].astype('category')
df['enginetype'] = df['enginetype'].astype('category')
df['fuelsystem'] = df['fuelsystem'].astype('category')
df['doornumber'] = df['doornumber'].astype('category')
df['cylindernumber'] = df['cylindernumber'].astype('category')
df['CarName'] = df['CarName'].cat.codes
df['fueltype'] = df['fueltype'].cat.codes
df['aspiration'] = df['aspiration'].cat.codes
df['carbody'] = df['carbody'].cat.codes
df['drivewheel'] = df['drivewheel'].cat.codes
df['enginelocation'] = df['enginelocation'].cat.codes
df['enginetype'] = df['enginetype'].cat.codes
df['fuelsystem'] = df['fuelsystem'].cat.codes
df['doornumber'] = df['doornumber'].cat.codes
df['cylindernumber'] = df['cylindernumber'].cat.codes
enc = OneHotEncoder()
enc_data = pd.DataFrame(enc.fit_transform(
    df[['CarName', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'fuelsystem', 'doornumber', 'cylindernumber']]).toarray())
New_df = df.join(enc_data)
X=df.drop(['price'], axis=1)
y=df['price']
print(X)
print(y)
```

```python
kf = KFold(n_splits=5, random_state=None)

results = []

lr = LinearRegression()

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    lr.fit(X_train, y_train)

    y_predLR = lr.predict(X_test)

    results.append(pd.DataFrame({'y_test': y_test, 'y_predict': y_predLR}))

resultDf = pd.concat(results)

print(resultDf)
```

```python
mean_absolute_error(y_test, y_predLR)
```
